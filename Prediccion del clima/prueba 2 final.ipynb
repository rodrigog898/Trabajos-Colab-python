```python
import pandas as pd
from sklearn.cluster import KMeans

# Supongamos que tienes un dataframe llamado "df" con las variables mencionadas

# Selecciona las columnas que deseas utilizar para el clustering
columns = ['RainTomorrow', 'Sunshine', 'WindGustSpeed', 'WindSpeed9am', 'WindSpeed3pm', 'Humidity3pm', 'Pressure9am', 'Pressure3pm', 'Cloud3pm', 'Temp9am', 'RISK_MM']
data = df_perth_5[columns]

# Crea una instancia del algoritmo K-medians con el número de clusters deseado
n_clusters = 3
kmedians = KMeans(n_clusters=n_clusters, init='random', algorithm='full')

# Ajusta el modelo a los datos
kmedians.fit(data)

# Obtiene los centroides y las etiquetas de los clusters
centroids = kmedians.cluster_centers_
labels = kmedians.labels_

# Agrega las etiquetas de los clusters al dataframe original
df_perth_5['Cluster'] = labels

# Muestra los resultados
print("Centroides:")
print(centroids)
print("\nEtiquetas de los clusters:")
print(labels)
```

```python

import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from kneed import KneeLocator
import matplotlib.pyplot as plt

# Supongamos que tienes un dataframe llamado df_n2 con las variables que mencionaste

# Seleccionamos las columnas relevantes para el análisis
data = df_n2[['MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation', 'Sunshine']]

# Normalizamos los datos
scaler = StandardScaler()
data_norm = scaler.fit_transform(data)

sse = []  # Guardaremos los valores de la función objetivo (SSE)

# Iteramos sobre diferentes valores de k
for k in range(1, 20):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data_norm)
    sse.append(kmeans.inertia_)

# Graficamos la curva de la función objetivo (SSE) en función de k
plt.plot(range(1, 20), sse, 'bx-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Sum of Squared Errors (SSE)')
plt.title('Elbow Method')
plt.show()

# Encontramos el número óptimo de clusters utilizando el método del codo
kl = KneeLocator(range(1, 20), sse, curve='convex', direction='decreasing')
optimal_k = kl.elbow

print("Número óptimo de clusters (k):", optimal_k)

```

```python
from minisom import MiniSom
import numpy as np
import matplotlib.pyplot as plt

# Supongamos que tienes un dataframe llamado df_n2 con las variables que mencionaste
data = df_n2[['MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation', 'Sunshine']]

# Normalizamos los datos
data_norm = (data - data.min()) / (data.max() - data.min())

# Definimos los parámetros de la red de Kohonen
n_rows = 10
n_columns = 10
input_len = data_norm.shape[1]
sigma = 1.0
learning_rate = 0.5
num_epochs = 100

# Creamos e inicializamos la red de Kohonen
som = MiniSom(n_rows, n_columns, input_len, sigma=sigma, learning_rate=learning_rate)
som.random_weights_init(data_norm.values)
som.train_random(data_norm.values, num_epochs)

# Obtenemos las coordenadas de los nodos ganadores
winners = np.array([som.winner(x) for x in data_norm.values])

# Graficamos los resultados
plt.figure(figsize=(8, 8))
plt.pcolor(som.distance_map().T, cmap='bone_r')
plt.colorbar()

# Agregamos los marcadores de los datos
for i, (x, y) in enumerate(winners):
    plt.text(x + 0.5, y + 0.5, str(i+1), color='r', ha='center', va='center', fontweight='bold')

plt.xticks(np.arange(n_columns+1))
plt.yticks(np.arange(n_rows+1))
plt.grid()
plt.title('Mapa de Kohonen')
plt.show()

```
